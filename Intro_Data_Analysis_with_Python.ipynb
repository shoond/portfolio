Simple Linear Regression is a linear appreach to model the relationship between a numerical target variable and one explanatory variable (predictors).

y = b + mx

Multiple Linear Regression is the most popular model for making predicitions. We fit a linear relationship between a numberical outcome (y) and a set of predictors (mx) to see which line fits best.

y = b1 + m1x1 + m2x2 + m3x3 + m4x4 + m5x5 +...

Variables in the linear equation must be numeric.

Non-integer attributes must be pre-processed into numeric values (e.g. Male = 0, Female = 1).

Partition the Data for Performance Evaluation
- Training Data and Validation Data

Data Representation for Data Mining
Machine learning is about creating models from data: for that reason, we’ll start by discussing how
data can be represented in order to be understood by the computer.

1. Attribute Matrix: a two-dimensional matrix that is most often contained in a NumPy
array or a Pandas DataFrame. The samples (i.e., rows) always refer to the individual objects
described by the dataset. For example, the sample might be a flower, a person, a document, an
applicant, a house, or anything else you can describe with a set of quantitative measurements.
The attributes (i.e., columns) always refer to the distinct properties that describe each sample
in a quantitative manner.
2. Target Array: The target array is usually one dimensional, and is generally contained in a
NumPy array.

1.2.1 Choose Predictors
Based on your business understanding and exploratory data analysis, choose predictors that might
have a (linear) impact on the target variable.
For example, we choose to use GROSS AREA, FLOORS and ROOMS as our predictors.

We
are looking for a linear regression model with unknown coefficients as


TOTAL VALUE = β0 + β1 × GROSS AREA + β2 × FLOORS + β3 × ROOMS.

```python
# Choose predictors to construct attribute matrix
select_index = ['GROSS AREA', 'FLOORS', 'ROOMS']
house_x = house.loc[:, select_index]

# Target
house_y = house.iloc[:,0]
```

1.3 Partition the Data into Train vs. Validation/Test
To create a train and validation set, we are going to split randomly this data set into 70% training
set and 30% validation/test set.

The random seed is set to 42.

We will use the sklearn.model_selection.train_test_split() function.

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = \
    train_test_split(house_x, house_y, test_size = 0.3, random_state = 42)
    # random_state allows for reproducability of randomized dataset
print(x_train)
```

1.4 Fit the Model to your Training Data


We first need to import the model from Scikit-Learn and then create the linear regression model
instance.

There are many other models available in Scikit-Learn. After you import the desired model, the
following pipelines and API’s are consistent. It can be easily changed from LinearRegression to
another model. For more information, please visit the documentation of Scikit-Learn. For example,
you can also do

e.g. from sklearn.naive_bayes import GaussianNB
model = GaussianNB()

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
import numpy as np

# Train the model using the training set
model.fit(x_train, y_train)

# Print out the fitted coeffcients
# np.set_printoptions() allows you to set precision

print(model.coef_)
print(model.intercept_)
```

1.5 Make Predictions for Any Data
Once the model is trained, the main task of regression is to evaluate it based on what it says about
new data that was not part of the training set. In Scikit-Learn, this can be done using the predict()
method.

For performance evaluation purposes, you need to get predictions for the validation set and com-
pare the predicted values to their true values. You can also make predictions for the training


set (by changing the input argument of the predict() from X_valid to X_train ), or any “new”
sample.
