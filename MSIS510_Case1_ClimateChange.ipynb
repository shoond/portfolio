# a_i
Build a linear regression model to predict Temp, using all other variables as predictors(except Year and Month). You should use the training set.

```python
# Create loop to concat independent vars
columns = [col for col in df_train_climate.columns if col!='Temp' and col!='Year' and col!='Month']

joiner = '+'.join(columns)

col_all = 'Temp ~ ' + joiner

print(col_all)
```

```python
# Fit linear reg model
Climate_linReg = smf.ols(col_all, data = df_train_climate)
Climate_linReg_results = Climate_linReg.fit()
print(Climate_linReg_results.summary())
```

# a_ii
Evaluate the quality of the model.
* What is R2
> R2 = 0.751
* Which independent vars are significant?
> It seems like all independent variables are significant except CH4 and N2O.

# b

Build new linear regression model, this time only using **MEI, TSI, Aerosols, and N2O** as independent variables. Remember to use training set to build model.


```python
# Create loop to concat independent vars
col_2 = [col for col in df_train_climate.columns if col=='MEI' or col=='TSI' or col=='Aerosols' or col=='N2O']

joiner = '+'.join(col_2)

col_2_all = 'Temp ~ ' + joiner

print(col_2_all)
```

```python
# Fit linear reg model
Climate_linReg_2 = smf.ols(col_2_all, data = df_train_climate)
Climate_linReg_results_2 = Climate_linReg_2.fit()
print(Climate_linReg_results.summary())
```

#b_i

How does the coefficient for N2O in this model compare to the previous model?

> N2O is positive compared to negative value of previous model indicating rise in N2O is a predictor of higher global temps.

#b_ii

How does the quality of this model compare to the previous one? Consider the R2 value and the significance of the independent vars when answering the question.

> R2 is slightly lower for this model, however all independent variables show as significant predictors of target variable, Temp.

# c

Using simple model from part b, calculate predictions for testing dataset.
* What is R2 on test set?
> 0.497
* What does this tell you about the model?
> The predictive variables we chose are roughly 50% probabilistic in predicting climate temperatures.


```python
temp_pred = Climate_linReg_results_2.predict(df_test_climate)

temp_test = df_test_climate['Temp']

temp_train_mean = np.mean(df_train_climate['Temp'])

# R2 = (SSE/SST)
# Find SSE and SST first
# SSE = SUM((obs - pred)**2)
# SST = SUM((obs-mean(pred))**2)
SSE_temp = np.sum(((temp_test - temp_pred)**2))

SST_temp = np.sum(((temp_test - temp_train_mean)**2))

R2 = 1 - (SSE_temp / SST_temp)

print(R2)


```
